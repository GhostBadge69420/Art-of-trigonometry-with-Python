```python
t = np.linspace(0, 10*np.pi, 150)
N = 100
 
numberOfVariations = 9
numberOfRowsAndCols = 3
 
fig, ax = plt.subplots(numberOfRowsAndCols , numberOfRowsAndCols, figsize=(15, 8))
row, col = np.indices((numberOfRowsAndCols, numberOfRowsAndCols))
 
# eds = exponential decays
eds = np.linspace(-6, -2, N)
for vari in range(numberOfVariations):
    spirals = np.linspace(1, vari+1, vari+1)
    r = row.ravel()[vari]
    c = col.ravel()[vari]
    ax[r, c].axis('off')
    ax[r,c].axis('equal')
    ax[r, c].set_title('No. of spirals = %s'%(vari+1))
        
    for turni, turnval in enumerate(spirals):
        for ni in range(N):
            x = np.cos(t+turnval) * np.exp(t/eds[ni])
            y = np.sin(t+turnval) * np.exp(t/eds[ni])
            ax[r, c].plot(x, y, color=[ni/N, ni/N, ni/N], linewidth=2, marker='1')
plt.show()
```

```python
# General initialization
N = 100 # number of lines to plot the each spiral
eds = np.linspace(-6, -2, N) # eds = exponential decays
resolutions = [10, 15, 20, 50, 100] # a list of resolutions for plotting the spirals
numberOfSpirals = [1, 2, 3, 4, 5, 6, 50, 100] # a list of variations for the number of spirals
 
# create the subplots
numberOfRows = len(resolutions) # number of rows for subplots
numberOfCols = len(numberOfSpirals) # number of columns for subplots
fig, ax = plt.subplots(numberOfRows , numberOfCols, figsize=(15, 8))
row, col = np.indices((numberOfRows, numberOfCols))
 
# Create and plot the graphs
for spirali, spiralval in enumerate(numberOfSpirals):
    spirals = np.linspace(1, spiralval, spiralval) # set the number of spirals for each graph  
    for resoli, resolval in enumerate(resolutions):
        t = np.linspace(0, 6*np.pi, resolval) # set the resolution for each graph
        
        #set the setting for plotting
        ax[resoli, spirali].axis("off")
        ax[resoli, spirali].axis("equal")
        ax[resoli, spirali].set_title("No. of spirals=%g\nresolution=%g" %(spiralval, resolval))
        for turni, turnval in enumerate(spirals):
            for ni in range(N):
                x = np.cos(t+turnval) * np.exp(t/eds[ni])
                y = np.sin(t+turnval) * np.exp(t/eds[ni])
                ax[resoli, spirali].plot(x, y, color=[ni/N, 0, ni/N], linewidth=2, marker='1')
plt.subplots_adjust(top=1.5) # output setting
plt.show()
```

```python
n = 100
t = np.linspace(0,20*np.pi,1000)
k = np.linspace(-3,-20,n)
for i in range(0,n):
    x = np.cos(t)*np.exp(t/k[i])
    y = np.sin(t)*np.exp(t/k[i])
    c = np.sin(i/n*np.pi)
    plt.plot(x,y,'-',color=[c,c*.3,c*.7],alpha=1)
plt.axis('square')
plt.gca().set_facecolor('black')
plt.xticks([])
plt.yticks([])
plt.show()
```
